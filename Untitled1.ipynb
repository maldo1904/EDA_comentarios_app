```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df['director'].fillna("Unknown",inplace=True)
df["cast"].fillna("Unknown",inplace=True)
df["country"].fillna("Unknown",inplace=True)
df.dropna(subset=["date_added"],inplace=True)
mas_comun=df["rating"].mode()[0]
```

```python
df.info()
```

```python
#¿Cuántos títulos se estrenaron por año?
print(df.groupby("release_year")["title"].count())
```

```python
import matplotlib.pyplot as plt
sns.histplot(x="release_year", data=df, discrete=True) # discrete=True para años
plt.title("Distribución de Títulos por Año de Lanzamiento")
plt.xlabel("Año de Lanzamiento")
plt.ylabel("Cantidad de Títulos")
plt.show()
```

Se puede observar que en los años más recientes se han lanzado más peliculas, es una tendencia muy alta

```python
#¿Cuál es el título con mayor duración?
index_max_duration = df["duration"].idxmax()
titulo_max_duracion = df.loc[index_max_duration, "title"]
max_duration = df.loc[index_max_duration, "duration"]

print(f"El titulo con mayor duracion es {titulo_max_duracion} (duración: {max_duration})")
```

```python
#¿Qué clasificaciones (Rating) son más comunes?
print(df["rating"].value_counts())
```

```python
sns.countplot(x="rating", data=df)
plt.xticks(rotation=90, ha='right')
plt.show()
```

TV-MA y TV-14 son notablemente los rating en mayor tendencia

```python
#¿Cómo se distribuyen los géneros (listed_in) más populares?
print(df["listed_in"].value_counts())
```

```python
genres = df["listed_in"].str.split(", ")

all_genres = pd.Series([genre for sublist in genres for genre in sublist])

all_genres.value_counts().plot(kind='bar', figsize=(10, 6))
plt.title("Distribución de Géneros Individuales")
plt.xlabel("Género")
plt.ylabel("Cantidad de Títulos")
plt.show()
```

International Movies es el genero más visto y TV shows es el menos visto

```python
#¿La duración promedio ha cambiado con los años?
import pandas as pd

def convert_duration_to_minutes(duration_str):
    if 'Season' in duration_str:
        # Asumimos una duración promedio por temporada (esto es una estimación)
        # Puedes ajustarla si tienes información más precisa
        return int(duration_str.split(' ')[0]) * 45  # Ejemplo: 45 minutos por temporada
    elif 'min' in duration_str:
        return int(duration_str.split(' ')[0])
    else:
        return None  # Manejar otros formatos si existen

df['duration_minutes'] = df['duration'].apply(convert_duration_to_minutes)

print(df.groupby("release_year")["duration_minutes"].mean())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

average_duration_by_year = df.groupby("release_year")["duration_minutes"].mean()

plt.figure(figsize=(10, 6))
sns.lineplot(x=average_duration_by_year.index, y=average_duration_by_year.values)
plt.title("Duración Promedio de Títulos por Año de Lanzamiento")
plt.xlabel("Año de Lanzamiento")
plt.ylabel("Duración Promedio (minutos)")
plt.grid(True)
plt.show()
```

Es muy irregular aquì ya que en los recientes años del df se puede observar que la media de la duración de las peliculas era muy corta poy lo cual esta se fue elevando mucho hasya alcanzar en aproximadamente 225 minutos de media en 1962 aproximadamente, despues empezo a bajar y tener algunos picos (no igual de altos), en la actualidad las peliculas son de corta duración (no como la de los primeros años del df pero si de aproximadamente de 65 minutos)

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.boxplot(x="release_year", y="duration_minutes", data=df)
plt.title("Distribución de la Duración de Títulos por Año de Lanzamiento")
plt.xlabel("Año de Lanzamiento")
plt.ylabel("Duración (minutos)")
plt.xticks(rotation=45, ha="right") # Rotar etiquetas si hay muchos años
plt.tight_layout()
plt.show()
```
